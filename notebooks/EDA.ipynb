Importing Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

Loading dataset

```python
#Reading the datasets and displaying the first 5 rows of each dataset
sales = pd.read_csv("../data/raw/sales.csv")
features = pd.read_csv("../data/raw/features.csv")
stores = pd.read_csv("../data/raw/stores.csv")
```

Dataset Overview

```python
print("SALES")
display(sales.head())

print("FEATURES")
display(features.head())

print("STORES")
display(stores.head())
```

```python
#Checking the shape of each dataset
print("Sales Shape:", sales.shape)
print("Features Shape:", features.shape)
print("Stores Shape:", stores.shape)
```

```python
#Checking the columns of each dataset
print("Sales Columns")
print(sales.columns.tolist())

print("\nFeatures Columns")
print(features.columns.tolist())

print("\nStores Columns")
print(stores.columns.tolist())
```

```python
#Checking the number of unique values in each object column of the sales dataset
for col in sales.select_dtypes(include="object"):
    print(col)
    print(sales[col].nunique())
    print("-"*40)
```

```python
#Checking the number of unique values in each object column of the features dataset
for col in features.select_dtypes(include="object"):
    print(col)
    print(features[col].nunique())
    print("-"*40)
```

```python
#Checking the number of unique values in each object column of the stores dataset
for col in stores.select_dtypes(include="object"):
    print(col)
    print(stores[col].nunique())
    print("-"*40)
```

```python
#Converting the date columns in sales and features datasets to datetime format
sales["date"] = pd.to_datetime(sales["date"])

features["date"] = pd.to_datetime(features["date"])
```

```python
sales.info() #checking the data types of the sales dataset after converting the date column to datetime format
```

```python
#Checking the start and end dates of the sales and features datasets
print("Sales Start :", sales["date"].min())
print("Sales End   :", sales["date"].max())

print("Features Start :", features["date"].min())
print("Features End   :", features["date"].max())
```
